El método enviar de la clase Envio se conecta directamente con la empresa de encomiendas y se encarga del despacho. No sabes cómo funciona este método, pero sabes que está implementado correctamente y funciona.


IMPORTANTE: NO DEBEN DEFINIR LA CLASE ENVIO.  Esta clase ya esta pre cargada en el código de la plataforma.

El definir la clase Envio nuevamente lleva a esta evaluación a su nota mínima.


Deberás definir las siguientes dos clases:


Carta

La clase Carta hereda de Envio. Esta clase tiene los siguientes métodos:

__init__(self, emisor, receptor): método que hace lo mismo que el __init__ de envío.

poner_estampilla(self): este método no recibe parámetros, pero al llamarse coloca una estampilla en la carta

enviar(self): este método deberá revisar que la carta tenga una estampilla y luego enviar correctamente con el método definido en Envio. Si la carta no tiene estampilla, entonces NO se deberá enviar.



Paquete

La clase Paquete hereda de Envio. Esta clase tiene los siguientes métodos:

__init__(self, emisor, receptor, max_volumen): método que deberá hacer lo mismo que el __init__ de envío, además de guardar max_volumen. Cuando se crea el paquete este está inicialmente abierto.

agregar_producto(self, peso, volumen): este método se llama cuando se quiere agregar un producto al paquete. Este método recibe como parámetros el peso y el volumen que tiene el producto a agregar a la caja. Si la caja está cerrada, entonces la caja se deberá abrir y agregar el producto.

cerrar(self): método que no recibe parámetro, este método se llama cuando se quiere cerrar el paquete. El paquete se deberá cerrar solamente si el volumen de todos los productos dentro del paquete es menos que el volumen máximo de la caja, en caso contrario la caja NO se deberá cerrar.

enviar(self): método que no recibe parámetros. Este método se llama cuando se quiere enviar el paquete. El método deberá enviar el paquete utilizando el método enviar de Envio solamente si el paquete está cerrado y el peso de todos los productos dentro del paquete es menor o igual a 30. En caso contrario NO se deberá enviar el paquete.


Puedes modificar todos los métodos mencionados anteriormente, sin embargo deben hacer como mínimo lo pedido en este enunciado.

````python
# RECORDATORIO:
# NO DEBEN DEFINIR LA CLASE ENVIO

class Envio:

    def __init__(self, emisor, receptor):
        self.emisor = emisor
        self.receptor = receptor

    def enviar(self):
        print(f"Enviado desde {self.emisor} a {self.receptor}")

# RECORDATORIO:
# NO DEBEN DEFINIR LA CLASE ENVIO

class Carta(Envio):
    def __init__(self, emisor, receptor):
        super().__init__(emisor, receptor)
        self.estampilla = False

    def poner_estampilla(self):
        self.estampilla = True

    def enviar(self):
        if self.estampilla:
            super().enviar()


class Paquete(Envio):

    def __init__(self, emisor, receptor, max_volumen):
        super().__init__(emisor, receptor)
        self.max_volumen = max_volumen
        self.productos = []
        self.cerrado = False

    def agregar_producto(self, peso, volumen):
        if self.cerrado:
            self.cerrado = False
        self.productos.append((peso, volumen))

    def cerrar(self):
        volumen_total = 0
        for producto in self.productos:
            volumen_total += producto[1]
        if volumen_total < self.max_volumen:
            self.cerrado = True

    def enviar(self):
        peso_total = 0
        for producto in self.productos:
            peso_total += producto[0]
        if self.cerrado and peso_total <= 30:
            super().enviar()

# 
c = Carta("Fred", "Black")
c.poner_estampilla()
c.enviar()  

# 
p = Paquete("Tienda", "Cliente", max_volumen=100)
p.agregar_producto(10, 30)
p.agregar_producto(5, 20)
p.cerrar()
p.enviar()  

# 
p2 = Paquete("Bodega", "Destino", max_volumen=40)
p2.agregar_producto(20, 25)
p2.agregar_producto(15, 20)
p2.cerrar()
p2.enviar()  
````
